```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## `plt.plot()`函数

```python
x = np.arange(100)
y1 = np.random.randn(100)

# 1. 绘制简单的线图
plt.plot(x, y1)
plt.show()

# 2. 指定线条颜色、样式和标记
plt.plot(x, y1, color = 'red', linestyle = '--', marker = 'o')
plt.show()

# 3. 绘制多条线
y2 = np.random.randn(100)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()

# 4. 添加标题和标签
plt.plot(x, y1)
plt.title('Sample Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()

# 5. 设置坐标轴范围
plt.plot(x, y1)
plt.xlim(20,80)
plt.ylim(-1,1)
plt.show()

# 6. 填充线图下方区域
plt.plot(x, y1)
plt.fill_between(x, y1, color = 'skyblue', alpha = 0.4)
plt.show()
```

```python
# 图片与子图
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.hist(np.random.randn(100), bins=20, color = 'k', alpha = 0.3)
ax2.hist()
ax3.plot(np.random.randn(50).cumsum(), 'k--')
plt.show()
```
